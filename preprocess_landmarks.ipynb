```python
#convert all lankmark files to PTS files
import os

path_to_landmarks = '/Programing/GR/Python/CK/landmarks/001'
path_to_images = '/Programing/GR/Python/CK/images/001'

landmarks = os.listdir(path_to_landmarks)
#print(landmarks)
for lm in landmarks:
    input = open(path_to_landmarks + '/' + lm, 'r')
    output = open(path_to_images + '/' + lm.replace('_landmarks.txt', '.pts'), 'w')
    output.write("version: 1" + '\n' + "n_points:  68" + '\n' + "{" + '\n')
    #output.write(input.readline());
    array = []
    for line in input: # read rest of lines
        array.append([float(x) for x in line.split()])
    for point in array:
        #print(str(point[0]) + ' ' + str(point[1])) 
        output.write(str(point[0]) + ' ' + str(point[1]) + '\n')
    output.write('}')
    output.close()
    input.close()
    
```

```python
import os

path_to_database = '/Programing/GR/CK+/cohn-kanade-images'
path_to_landmarks = '/Programing/GR/CK+/Landmarks'

landmarks_folder = os.listdir(path_to_landmarks)
for folder in landmarks_folder:
    #print(os.listdir(path_to_landmarks +'/'+ folder))
    sub_folder = os.listdir(path_to_landmarks +'/'+ folder)
    for sf in sub_folder:
        images_dir = path_to_database + '/' + folder + '/' + sf
        landmarks_dir = path_to_landmarks + '/' + folder + '/' + sf
        #print(landmarks_dir)
        landmarks = os.listdir(landmarks_dir)
        for lm in landmarks:
            if(lm != '.DS_Store'):
                input = open(landmarks_dir + '/' + lm, 'r')
                output = open(images_dir + '/' + lm.replace('_landmarks.txt', '.pts'), 'w')
                output.write("version: 1" + '\n' + "n_points:  68" + '\n' + "{" + '\n')
                array = []
                for line in input: # read rest of lines
                    array.append([float(x) for x in line.split()])
                for point in array:
                    #print(str(point[0]) + ' ' + str(point[1])) 
                    output.write(str(point[0]) + ' ' + str(point[1]) + '\n')
                output.write('}')
                output.close()
                input.close()

```

```python
import os
import glob
from shutil import copyfile

#tao database cho landmark aam training - chi lay file .png dau va cuoi cua moi sequence

path_to_database = '/Programing/GR/Code/CK+/cohn-kanade-images'
path_to_aam_database = '/Programing/GR/Code/CK+/aam-images'

landmarks_folder = os.listdir(path_to_database)
#print(landmarks_folder)
for folder in landmarks_folder:
    sub_folder = os.listdir(path_to_database +'/'+ folder)
    #print(sub_folder)
    for sf in sub_folder:
        directory = '/' + folder + '/' + sf
        file_list = os.listdir(path_to_database + directory)
        if(file_list[-1] == "Thumbs.db"):
            file_list.remove("Thumbs.db")
        #print(file_list[0])
        if not os.path.exists(path_to_aam_database + directory):
            os.makedirs(path_to_aam_database + directory)
        src = path_to_database + directory + '/' + file_list[0]
        dst = path_to_aam_database + directory + '/' + file_list[0]
        copyfile(src, dst)
        src = path_to_database + directory + '/' + file_list[1]
        dst = path_to_aam_database + directory + '/' + file_list[1]
        copyfile(src, dst)
        src = path_to_database + directory + '/' + file_list[-1]
        dst = path_to_aam_database + directory + '/' + file_list[-1]
        copyfile(src, dst)
        src = path_to_database + directory + '/' + file_list[-2]
        dst = path_to_aam_database + directory + '/' + file_list[-2]
        copyfile(src, dst)
```
